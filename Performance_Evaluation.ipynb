```python
from sklearn import datasets

iris_data = datasets.load_iris()
```

```python
X = iris_data['data']

y = iris_data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
model = DecisionTreeClassifier()

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
model = DecisionTreeClassifier(max_depth=2)

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# HW6 วนลูปทำ 10-fold cross validation

- search : Generate code for training decision charts and evaluate them with 10-fold cross-validation, outputting data one fold at a time.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold

# Load the iris dataset
iris = load_iris()
X = iris_data['data']
y = iris_data['target']
```

```python
# Define the decision tree classifier with desired hyperparameters
dt = DecisionTreeClassifier(max_depth=3)

# Define the 10-fold cross-validation object
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using cross-validation
fold_num = 1
for train_index, test_index in kf.split(X):
    print(f"Fold {fold_num}:")

    # Split the data into training and testing sets for this fold
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Fit the model on the training set
    dt.fit(X_train, y_train)

    # Evaluate the model on the testing set and print the results
    accuracy = dt.score(X_test, y_test)
    print(f"Accuracy: {accuracy:.3f}\n")

    fold_num += 1
```
